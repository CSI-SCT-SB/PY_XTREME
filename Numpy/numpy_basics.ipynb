```python
#VISHNU R

import numpy as np
arr=np.array([1,2,3,'a'])
print(arr)
print(type(arr))

arr2=np.linspace(start=1,stop=6,num=15,endpoint=True,retstep=False) # create random array
print(arr2)
print(type(arr2))
print(np.shape(arr2))   #to print dimension of array
print(np.shape(arr))

arr3=np.arange(start=1,stop=11,step=3) #create array with difference 3
print(arr3)

arr4=np.ones((2,2))  #create array with all 1s
print(arr4)

arr5=np.zeros((3,2)) #array with all 0s
print("\n",arr5)

arr6=np.random.rand(1,2)
print(arr6)

arr7=np.logspace(start=1,stop=10,num=5,base=10)  #create a log valued array
print(arr7)
```
